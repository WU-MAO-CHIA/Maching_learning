```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
x = pd.period_range(pd.datetime.now(), periods= 200, freq= 'd')
# 建造一連串的時間
x = x.to_timestamp().to_pydatetime() #時間轉時間戳記, 
y = np.random.randn(200, 3).cumsum(0)
plt.plot(x, y)
plt.show()
```

```python
# matplotlib 用點 point 而非 pixel 為圖尺寸的測量單位， 1 point = 1/ 72 inchs
mpl.rcParams['lines.linewidth'] = 5 #set  lines width
mpl.rcParams['lines.color'] = 'r' # set line color
mpl.rcParams['figure.figsize'] = (10, 10) # 設定畫布大小
plt.gcf().set_size_inches(10, 10) # 和figure.figsize 相同但用 inches 為單位

x = pd.period_range(pd.datetime.now(), periods= 200, freq= "d")
x = x.to_timestamp().to_pydatetime() #
y = np.random.randn(200, 3).cumsum(0)
plt.plot(x, y)
plt.show()
```

```python
plots = plt.plot(x, y)
plt.legend(plots, ('apple', 'facebook', 'google'), loc= 'best', framealpha= 0.5, prop= {'size': 'large', 'family':'monospace'})
plt.show()
# loc : 0: ‘best'1: ‘upper right'2: ‘upper left'3: ‘lower left'4: ‘lower right'
#       5: ‘right'6: ‘center left'7: ‘center right'8: ‘lower center'9: ‘upper center'10: ‘center'
#prop : 字體參數
```

```python
plt.title('Random Trends')
plt.xlabel('Date')
plt.ylabel('Cum. Sum')
plt.figtext(0.995, 0.01, 'CopyRight', ha= 'left', va= 'top')
plt.tight_layout()
plt.plot(x, y)
plt.show()
```

```python
fig = plt.figure(figsize= (8, 4), dpi= 200, tight_layout= True, linewidth= 1, edgecolor= 'r')
```

```python
fig = plt.figure(figsize= (8, 4))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.set_title('Main Axes with Insert child Axes')
ax.plot(x, y[:, 0])
ax.set_xlabel('Date')
ax.set_ylabel('Cum. sum')

ax = fig.add_axes([0.15, 0.15, 0.3, 0.3])
ax.plot(x, y[:, 1], color= 'r')
ax.set_xticks([])
```

```python
figure, ax = plt.subplots()
plots = ax.plot(x, y, label= '')
figure.set_size_inches(8, 4)
ax.legend(plots, ('apple', 'facebook', 'google'), loc= 'best', framealpha= 0.25, prop= {'size': 'small', 'family': 'monospace'})
# framealpha 是控制圖例框架的自由度

ax.set_title('Random trends')
ax.set_xlabel('Date')
ax.set_ylabel('Cum. sum')

ax.grid(True) #背景有格子
figure.text(0.995, 0.01, "ACM 2015", ha= 'left', va= 'bottom')
figure.tight_layout()
```

```python
figure, ax = plt.subplots()
plots = ax.plot(x, y, label= '')
figure.set_size_inches(8, 4)
ax.legend(plots, ('apple', 'facebook', 'google'), loc= 'best', framealpha= 0.25, prop= {'size' : 'small', 'family' : 'monospace'})
ax.set_xlabel("Date")
ax.set_ylabel('Cum .sum')
ax.grid(True)

figure.text(0.995, 0.01, 'Acm', ha= 'right', va= 'bottom')
figure.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=3, ncols= 1, sharex= True, sharey= True, figsize= (8, 8))
labelled_data = zip(y.transpose(), ('Apple', 'Facebook', 'google'), ('b', 'g', 'r'))
fig.suptitle('Three Random Trends', fontsize= 16)

for i, ld in enumerate(labelled_data):
    ax = axes[i]
    ax.plot(x, ld[0], label= ld[1], color= ld[2])
    ax.set_ylabel('Cum. sum')
    ax.legend(loc= 'upper left', framealpha= 0.5, prop= {'size':'small'})
    
axes[-1].set_xlabel('Date')
```
