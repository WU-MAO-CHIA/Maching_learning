```python
import torch
```

```python
x = torch.ones(5)
y = torch.zeros(3)
w = torch.randn(5, 3, requires_grad= True)
b = torch.randn(3, requires_grad= True)
z = torch.matmul(x, w) + b
loss = torch.nn.functional.binary_cross_entropy_with_logits(z, y)
print(w)
print(b)
```

```python
print(f'Gradient function for z = {z.grad_fn}')
print(f'Gradient function for loss = {loss.grad_fn}')
```

```python
loss.backward()
print(w.grad)
print(b.grad)
```

```python
z = torch.matmul(x, w)+ b
print(z.requires_grad)

with torch.no_grad():
  z = torch.matmul(x, w) + b
print(z.requires_grad)
```

```python
z = torch.matmul(x, w) + b
z_det = z.detach() #same with the torch.no_grad
print(z_det.requires_grad)
```

```python
inp = torch.eye(5, requires_grad= True) #torch.eye 建立一個像單位矩陣的tensor
print(f'inp\n{inp}\n')
out = (inp + 1).pow(2)
print(f'out\n{out}\n')
out.backward(torch.ones_like(inp), retain_graph= True) #微分一次的係數
print(f'First call\n{inp.grad}') 
out.backward(torch.ones_like(inp), retain_graph= True) #微分兩次的係數和
print(f'\nSecond call\n{inp.grad}')
inp.grad.zero_() #重設微分和
out.backward(torch.ones_like(inp), retain_graph= True)
print(f'\ncall after zeroing gradients\n{inp.grad}')
```
