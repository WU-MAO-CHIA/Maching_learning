```python
import torch
import numpy as np
```

```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
```

```python
np_array = np.array([[1, 2], [3, 4]])
x_np = torch.from_numpy(np_array)
```

```python
x_ones = torch.ones_like(x_data, dtype = torch.float)
print(f'Ones Tensor: \n {x_ones} \n')

x_rand = torch.rand_like(x_data, dtype= torch.float)
print(f'Random Tensor: \n {x_rand} \n')
```

```python
shape = (2, 3,) #最後一個逗點可以省略
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f'Random Tensor: \n {rand_tensor} \n')
print(f'Ones Tensor: \n {ones_tensor} \n')
print(f'Zeros Tensor: \n {zeros_tensor}')
```

```python
tensor = torch.rand(3, 4)
print(f'shape of tensor: {tensor.shape}')
print(f'datatype of tensor: {tensor.dtype}')
print(f'device tensor is stored on: {tensor.device}')
```

```python
tensor = torch.ones(4, 4)
tensor[:, 0] = 0
tensor[:, 1] = 1
tensor[:, 2] = 2
tensor[:, 3] = 3
print(f'first row: {tensor[0]}')
print(f'first column: {tensor[:, 0]}')
print(f'last column: {tensor[..., -1]}') # ... 

print(tensor)
```

```python
t1 = torch.cat([tensor, tensor, tensor], dim= 1)
t2 = torch.cat([tensor, tensor, tensor], dim= 0)
print(t1)
print(t2)
```

```python
y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)
print(tensor.T, '\n')
print(y1, '\n')
print(f'{y2} \n ')
y3 = torch.rand_like(tensor)
torch.matmul(tensor, tensor.T, out= y3)
print(y3, '\n')
z1 = tensor * tensor
z2 = tensor.mul(tensor)

z3 = torch.rand_like(tensor)
torch.mul(tensor, tensor, out= z3)
```

```python
agg = tensor.sum()
agg_item = agg.item()
print(agg, agg_item, type(agg), type(agg_item))
```

```python
np_array = np.array([[1, 2], [3, 4]])
x_np = torch.from_numpy(np_array)

np_array = np.array([[5, 6], [7, 8]])
y_np = torch.from_numpy(np_array)

print(x_np)
x_np.copy_(y_np)
print(x_np)
```

```python
x = torch.tensor(np.array([[1, 2], [3, 4]]))
z = x.add(5)
print(z,'\n', x)
y = x.add_(5)
print(y, '\n', x)
```

```python
t = torch.ones(5)
print(f't: {t}', type(t))
n = t.numpy()
print(f'n: {n}', type(n))
```

```python
t.add_(1)
print(f't: {t}')
print(f'n: {n}')
# t.numpy() just like a pointer to tensor
```

```python
print(f'{tensor} \n')
tensor.add_(5)
print(tensor)
```

```python
t = torch.ones(5)
print(f't: {t}')
n = t.numpy()
print(f'n: {n}')
t.add_(1)
print(t)
print(n)
```

```python
n =  np.ones(5)
t = torch.from_numpy(n)
np.add(n, 1, out= n)
print(f't: {t}')
print(f'n: {n}')
# 不論從numpy 轉 tensor 或 tensor 轉 numpy 都 share same memory location
# 所以只要改變一個另一個就會跟著改變
```
